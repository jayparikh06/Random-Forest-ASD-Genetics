```python
df_3 = pd.read_csv("/content/SFARILevel3.csv")
df_3 = df_2.fillna(0)
```

```python
# df_2 and df_3 are our inference dataframes
final_inference_df = pd.DataFrame(columns=df_train.columns)

inf_datasets = [df_2, df_3]
# for each of the safari 2 and 3 datasets
for dataset in inf_datasets:
  # for each index in the dataset we chose
  for index in range(len(dataset)):
    # get the row at that index and convert it to a dictionary
    cur_row = dataset.iloc[index].to_dict()
    # start the dictionary that represents our new row that we add to our final inference df
    new_rowdict = {"Genes": cur_row["Gene Name"]}
    # iterate through the columns of the original train dataset
    for item in df_train.columns:
      # we already have this, so ignore
      if item == "Genes":
        continue

      # if the column is in our current safari 2 or 3 dataset, add in the dataset value
      if item in cur_row.keys():
        new_rowdict[item] = cur_row[item]
      # if not, set it to 0
      else:
        new_rowdict[item] = 0.0
    final_inference_df = final_inference_df.append(new_rowdict, ignore_index=True)

final_inference_df.head()
```
